```python
# Import libraries
import pandas as pd
#from pandas import ExcelWriter
#from pandas import ExcelFile
#import datetime
#import matplotlib as plt
import numpy as np
```

## Jax forecast clean up

```python
# Load jax fc file
jax_fc_file = "Resources/JAX FC EXCEL.xlsx"

# Read Excel file into DataFrame, set Item as index
#jax_fc = pd.read_csv(jax_fc_file, index_col = "Item", parse_dates=["Date"])
jax_fc = pd.read_excel(jax_fc_file, index_col = "Item")
jax_fc.head()
```

```python
# Remove extra columns 
jax_fc = jax_fc[["Date", "Current", "Customer", "Price"]]
jax_fc.head()
```

```python
# Change Current to Qty
jax_fc = jax_fc.rename(columns = {"Current": "Qty"})
jax_fc.head()
```

```python
# Remove Qty 0
jax_fc = jax_fc[jax_fc["Qty"] != 0]
jax_fc.head()
```

```python
#Check data types
jax_fc.dtypes
```

```python
# Count of items
jax_fc.count()
```

```python
# Add extended price column and populate with Price * Qty
jax_fc["Extended Price"] = jax_fc.Price * jax_fc.Qty
jax_fc.head()
```

```python
# Add column TYPE
jax_fc["Type"] = "FC"
jax_fc.head()
```

```python
# Add column Org
jax_fc["Org"] = "JAX"
jax_fc.head()
```

```python
# New dataframe with updated forecast info
final_jax_fc = jax_fc
final_jax_fc
```

```python
# Check total count against resource file to ensure data is correct
final_jax_fc.count()
```

```python
# Sort by Date, then Item 
final_jax_fc = final_jax_fc.sort_values(["Item", "Date"])
final_jax_fc.head()
```

```python
# Save file to csv
final_jax_fc.to_csv("output_data/final_jax_fc_2.csv")
```

```python
# Save to Excel
final_jax_fc.to_excel("output_data/final_jax_fc_2.xlsx")
```
